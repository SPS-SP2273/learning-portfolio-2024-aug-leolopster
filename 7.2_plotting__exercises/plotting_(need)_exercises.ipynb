**Exercise 1 (Fill in the blanks)** Fill in the blanks(indicated with a ?), so the code below will produce the plot shown.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='darkred', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='k', capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

**Exercise 2 (A simple mathematical plot)**

Use NumPy to generate x and y data suitable for plotting the graph of the function:
$$
y = e^{-x/10}sin(x)
$$

for values of x in the range 0 to 50.

```python
x = np.linspace(0, 50, 250)
y = np.exp(-x/10) * np.sin(x)
```

```python
plt.plot(x, y)
plt.rcParams['text.usetex'] = True
plt.title(r'$y = e^{-x/10}\sin(x)$', fontsize=20)
plt.xlabel('x', color='orange', fontsize=15)
plt.ylabel('y', color='orange', fontsize=15, rotation='horizontal')
plt.ylim((-1, 1))
plt.grid(alpha=.25)
plt.show()
```
