1. Verify that both codes give the same result (HINT: print()).

```python
squares1 = [i**2 for i in range(10)]
squares2 = list()
for i in range(10):
    squares2.append(i**2)
print(squares1, squares2)
```

2. Use %%timeit to measure the execution time for each example.

```python
%%timeit
squares1 = [i**2 for i in range(10)]
```

```python
%%timeit
squares2 = list()
for i in range(10):
    squares2.append(i**2)
```
